```python
sales['month_year'] = sales['Order Date'].dt.to_period('M')
```

```python
monthly_profit = sales.groupby('month_year')['Total Profit'].sum().reset_index()
```

```python
monthly_profit = sales.sort_values(by ='Total Profit',ascending=False).head(10)
```

```python
monthly_profit['month_year'] = monthly_profit['month_year'].astype(str)
```

```python
mlt.figure(figsize=(10, 6))
mlt.plot(monthly_profit['month_year'], monthly_profit['Total Profit'], marker='o')
mlt.title('Total Profit by Month-Year')
mlt.show()
```
